```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [4, 5, 6, 7, 8, 9, 10]
y_brute_force = [0, 0, 1, 2, 5, 129, 205]
y_parallel_1 = [2, 3, 4, 9, 15, 82, 110]
y_parallel_2 = [2, 2, 2, 3, 4, 137, 142]
y_parallel_3 = [60, 63, 66, 71, 70, 145, 160]
```

```python
plt.plot(x, y_brute_force, label='Brute Force')
plt.plot(x, y_parallel_1, label='Parallel 1 (OpenMP)')
plt.plot(x, y_parallel_2, label='Parallel 2 (MPI)')
plt.plot(x, y_parallel_3, label='Parallel 3 (OpenMP + MPI)')
plt.xlabel('Number of elements')
plt.ylabel('Time (ms)')
plt.legend()
plt.show()
```

```python
y_brute_force_11_nodes = 34588
y_parallel_1_11_nodes = 18236
y_parallel_2_11_nodes = 27666
y_parallel_3_11_nodes = 19734
```

```python
# Plot bar plot
labels = ['Brute Force', 'Parallel 1 (OpenMP)', 'Parallel 2 (MPI)', 'Parallel 3 (OpenMP + MPI)']
y = [y_brute_force_11_nodes, y_parallel_1_11_nodes, y_parallel_2_11_nodes, y_parallel_3_11_nodes]
x = np.arange(len(labels))

plt.figure(figsize=(10, 5))
plt.title('Execution time for 11 nodes')
plt.bar(x, y)
plt.xticks(x, labels)
plt.ylabel('Time (ms)')
plt.show()
```
