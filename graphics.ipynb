```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [4, 5, 6, 7, 8, 9, 10]
y_brute_force = [0, 0, 1, 2, 5, 129, 205]
y_parallel_1 = [2, 3, 4, 9, 15, 82, 110]
```

```python
plt.plot(x, y_brute_force, label='Brute Force')
plt.plot(x, y_parallel_1, label='Parallel 1')
plt.xlabel('Number of elements')
plt.ylabel('Time (ms)')
plt.title('Brute Force vs Parallel 1')
plt.legend()
plt.show()
```
